```python
#droping all the columns with null values more than 50%
df2 = df[[col for col in df if df[col].count()/len(df)>=0.5]]
df2.head()
```

```python
#dropping all the unwanted columns
df2.drop(['Photo','Flag','Club Logo'],axis=1,inplace=True)

df2.drop('Unnamed: 0',axis=1,inplace=True)

df2.shape
```

```python
df2.head()
```

```python
#data cleaning
df2['Value'] = pd.to_numeric(df2['Value'].apply(lambda x : x.replace('€','').replace('M','').replace('K','')))

df2['Wage'] = pd.to_numeric(df2['Wage'].apply(lambda x : x.replace('€','').replace('M','').replace('K','')))

df2['Release Clause'] = pd.to_numeric(df['Release Clause'].apply(lambda x : x.replace('€','').replace('M','').replace('K','') if type(x) == str else x))
```

##### converting to category

```python
club = df2.Club.unique()
C ={}
j = 0
for i in club:
    C[i] = j
    j += 1
c = {v:k for k,v in C.items()}
df2.Club = df2.Club.map(C)
```
