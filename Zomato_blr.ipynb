```python
import os
import sys
```

```python
path = 'F:\machine learning and deep learning\datasets\zomato-india-data-set\Bengaluru\\'
```

```python
#list of all the file present the directory
a = os.listdir(path)

b = a[0]
del a[0]
```

```python
#reading the first file
import numpy as np
import pandas as pd
data_master = pd.read_csv(path+b,sep="|")
data_master.head()
data_master.shape
```

```python
#concating all the files
for i in a:
    df = pd.read_csv(path+i,sep="|")
    data_master = pd.concat([data_master,df])
data_master.shape
```

```python
data_master.info()
```

```python
data_master.head()
```

```python
data_master.columns
```

```python
#removing all the unwanted columns
data_master.drop(['CITY','URL','PAGE NO','TIMING','NAME'],axis=1,inplace=True)
data_master.head()
```

```python
data_master.info()
```

```python
#rearranging the inddex of the dataframe
data_master.index = range(data_master.shape[0])
```

```python
data_master.CUSINE_CATEGORY[1001].strip().split(',')
```

```python
#converting all the category tree to different columns
li = []
CAT = {}
for i in data_master.index:
    li.append(str(data_master.CUSINE_CATEGORY[i]).strip().split(','))
    CAT[i] = li[i]
CAT
```

```python
data_master.head()
```

```python
#checking for how many rows are unique in the category
data_master['CUSINE_CATEGORY'].nunique()
```

```python
#checking the maximum and minimum number of category in the category tree
le = []
for i in li:
    le.append(len(i))
print(max(le),min(le))
```

```python
#padding the category tree to the maximum number as above
level = []
for i in CAT.values():
    if len(i) < 8:
        for j in range(8-len(i)):
            i.append('')
    level.append(len(i))
```

```python
#creating another dataframe for category
cat_df = pd.DataFrame(CAT).T
print(cat_df.shape)
cat_df.head(10)
```

```python
#renaming the columns
cat_df.columns = ['category','sub-cat1','sub-cat2','sub-cat3','sub-cat4','sub-cat5','sub-cat6','sub-cat7']
```

```python
cat_df.head()
```

```python
#merging the datafreames
data_master = data_master.merge(cat_df, left_index=True, right_index=True)
data_master.head()
```

```python
#dropping the category tree as we don't need it anymore
data_master.drop('CUSINE_CATEGORY',axis=1,inplace=True)
```
